## Specifying domains and coordinate reference systems

earthkit.maps offers a variety of methods for setting the domain and/or coordinate reference system (CRS) of a chart.

```python
import earthkit.maps
```

### Global charts

If no domain is set when creating a new chart, earthkit.maps will assume that the domain is **global**. By default, it will use an equirectangular (Plate Carr√©e) CRS.

```python
chart = earthkit.maps.Map()
chart.coastlines()
chart.gridlines()
chart.show()
```

A coordinate reference system can be specified with the `crs` argument to the `earthkit.maps.Map` constructor. This can be a string or a cartopy CRS - see the cartopy documentation for a [list of available options](https://scitools.org.uk/cartopy/docs/latest/reference/projections.html?highlight=projections).

```python
chart = earthkit.maps.Map(crs="Robinson")
chart.coastlines()
chart.gridlines()
chart.show()
```

```python
import cartopy.crs as ccrs

chart = earthkit.maps.Map(crs=ccrs.Robinson())
chart.coastlines()
chart.gridlines()
chart.show()
```

```python
chart = earthkit.maps.Map(crs="InterruptedGoodeHomolosine")
chart.coastlines()
chart.gridlines()
chart.show()
```

### Named domains

Domains can also be set using the name of a country, continent or state/province. The named domain must exist in [Natural Earth's administrative regions shapefiles](https://www.naturalearthdata.com/downloads/50m-cultural-vectors/), which are used to get a bounding box for the named area.

```python
chart = earthkit.maps.Map(domain="Europe")
chart.coastlines()
chart.gridlines()
chart.show()
```

```python
chart = earthkit.maps.Map("Asia")
chart.coastlines()
chart.gridlines()
chart.show()
```

```python
chart = earthkit.maps.Map(domain="Antarctica")
chart.coastlines()
chart.gridlines()
chart.show()
```

```python
chart = earthkit.maps.Map(domain="Alaska")
chart.coastlines()
chart.gridlines()
chart.show()
```

```python
chart = earthkit.maps.Map(domain="Greenland")
chart.coastlines()
chart.gridlines()
chart.show()
```

It's also possible to pass **both** a domain **and** a CRS to a earthkit.maps chart, which will set the best bounding box for the named domain on the CRS of choice.

```python
chart = earthkit.maps.Map(domain="Greenland", crs=ccrs.Robinson())
chart.coastlines()
chart.gridlines()
chart.show()
```

### Bounding boxes

The other option for defining a chart's domain is a four-element bounding box containing the latitude and longitude bounds of the rectangle to plot. These should be provided in the order `[min_lon, max_lon, min_lat, max_lat]`.

earthkit.maps will attempt to generate an optimal CRS for the provided bounding box, but you still have the option of passing a custom CRS if desired.

```python
chart = earthkit.maps.Map(domain=[-20, 40, 32, 72])
chart.coastlines(resolution="50m")
chart.gridlines()
chart.show()
```

```python
chart = earthkit.maps.Map(domain=[-20, 40, 32, 72], crs=ccrs.NorthPolarStereo(central_longitude=7.5))
chart.coastlines(resolution="50m")
chart.gridlines()
chart.show()
```
