```python
import earthkit
```

```python
data = earthkit.data.from_source(
    'cds', 'reanalysis-era5-single-levels',
    {
        'product_type': 'reanalysis',
        'variable': '2m_temperature',
        'year': '2013',
        'month': '03',
        'day': '25',
        'time': '00:00',
    },
)
```

```python
style = earthkit.maps.styles.Style(colors="turbo", levels=range(-40, 41))
```

```python
chart = earthkit.maps.Chart(crs="Mollweide")
chart.shaded_contour(data, style=style, units="celsius", extend="both")
chart.coastlines()
chart.ocean(color="#333", zorder=2)
chart.gridlines(draw_labels=False, xlocs=range(-180, 181, 20), ylocs=range(-90, 90, 10))
chart.legend(location="top")
chart.show()
```
