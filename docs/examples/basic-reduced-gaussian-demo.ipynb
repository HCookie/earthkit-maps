```python
import earthkit
from earthkit.data import remote_example_file
```

```python
data = earthkit.data.from_source("url", remote_example_file("redgaus.grib"))
data.ls()
```

```python
chart = earthkit.maps.Chart()
chart.tricontourf(data, units="celsius")
chart.coastlines()
chart.gridlines()
chart.legend()
chart.show()
```
