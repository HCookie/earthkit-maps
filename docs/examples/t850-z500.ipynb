```python
import earthkit
```

```python
z500 = earthkit.data.from_source(
    'cds',
    'reanalysis-era5-pressure-levels',
    {
        'product_type': 'reanalysis',
        'variable': 'geopotential',
        'pressure_level': '500',
        'year': '2023',
        'month': '08',
        'day': '16',
        'time': '12:00',
    },
)

t850 = earthkit.data.from_source(
    'cds',
    'reanalysis-era5-pressure-levels',
    {
        'product_type': 'reanalysis',
        'variable': 'temperature',
        'pressure_level': '850',
        'year': '2023',
        'month': '08',
        'day': '16',
        'time': '12:00',
    },
)
```

```python
earthkit.maps.quickplot(t850, z500, units=("celsius", "dam"))
```

```python
chart = earthkit.maps.Superplot()

chart.plot(t850, units="celsius")
chart.plot(z500, units="dam")

chart.coastlines()
chart.gridlines()

chart.legend()

chart.title("T850 and Z500 on {time}")

chart.show()
```
