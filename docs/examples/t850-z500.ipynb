```python
import earthkit
```

```python
z500 = earthkit.data.from_source(
    'cds',
    'reanalysis-era5-pressure-levels',
    {
        'product_type': 'reanalysis',
        'variable': 'geopotential',
        'pressure_level': '500',
        'year': '2023',
        'month': '08',
        'day': '16',
        'time': '12:00',
    },
)

t850 = earthkit.data.from_source(
    'cds',
    'reanalysis-era5-pressure-levels',
    {
        'product_type': 'reanalysis',
        'variable': 'temperature',
        'pressure_level': '850',
        'year': '2023',
        'month': '08',
        'day': '16',
        'time': '12:00',
    },
)
```

```python
style = earthkit.maps.styles.Contour(
    colors= [
        (0, 0, 0.5),
        (0, 0, 0.5),
        (0, 0, 0.5),
        (0, 0, 0.5),
        (0, 0, 0.5),
        (0, 0, 0.85),
        (0.25, 0, 1),
        (0.5, 0, 1),
        (0, 0.5, 1),
        (0, 1, 1),
        (0, 1, 0.5),
        (0.5, 1, 0),
        (0.8536, 1, 0),
        (1, 1, 0),
        (1, 0.9619, 0),
        (1, 0.8536, 0),
        (1, 0.6913, 0),
        (1, 0.5, 0),
        (1, 0.3087, 0),
        (1, 0.1464, 0),
        (1, 0.0381, 0),
        (1, 0, 0),
        (1, 0, 1),
        (1, 0, 1),
        (1, 0, 1),
        (1, 0, 1),
        (1, 0, 1),
    ],
    levels=range(-48, 57, 4),
    units="celsius",
)
```

```python
chart = earthkit.maps.Superplot()

chart.plot(t850, style=style)
chart.plot(z500, units="dam")

chart.coastlines()
chart.gridlines()

chart.legend()

chart.subplot_titles("T850 and Z500 on {time}")

chart.show()
```
