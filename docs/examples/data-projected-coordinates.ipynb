```python
import earthkit
```

```python
efas = earthkit.data.from_source("url", earthkit.data.remote_example_file("efas.nc"))[0]
efas
```

```python
efas.projection().to_cartopy_crs()
```

```python
discharge_style = earthkit.maps.styles.Style(
    colors="Blues",
    levels=[0.1, 0.5, 1, 2, 5, 10, 50, 100, 500, 1000, 2000, 3000, 4000],
    normalize=True,
)
```

```python
chart = earthkit.maps.Chart(crs=efas.projection())
chart.shaded_contour(efas, style=discharge_style)
chart.coastlines()
chart.gridlines()
chart.borders()
chart.legend()
chart.title()
chart.show()
```

```python
import cartopy.crs as ccrs

chart = earthkit.maps.Chart(crs=ccrs.NorthPolarStereo(central_longitude=10))
chart.shaded_contour(efas, style=discharge_style, extend="max", transform_first=True)
chart.coastlines(resolution="medium")
chart.gridlines()
chart.borders()
chart.legend()
chart.title()
chart.show()
```
