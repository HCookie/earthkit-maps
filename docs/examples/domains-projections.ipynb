# Domains and projections

```python
from earthkit.maps import Chart, quickplot
```

Images in `earthkit-maps` are generated on a `Chart` object.

- `crs`: The coordinate reference system to use for the chart. This can be a cartopy `CRS` object, or a string name of a coordinate reference system.
- `domain`: The name of the domain on which you would like to plot your chart. This can be the name of a country, a continent, or a list of domain extents.

```python
chart = Chart(domain="Europe")
chart.coastlines()
chart.land()
chart.borders()
chart.gridlines()
chart.show()
```

The `domain` and `crs` arguments can also be passed to `earthkit.maps.quickplot`, so for brevity we will use `quickplot` for the remaining examples.

If you are running this example in an interactive Jupyter session, you should be able to play with various domains with widgets below.

### Continents

```python
import math

CONTINENTS = [
    "Europe",
    "Asia",
    "Africa",
    "North America",
    "South America",
    "OCeania",
]

chart = Chart(cols=3, rows=math.ceil(len(CONTINENTS)/4))

for continent in CONTINENTS:
    chart.add_subplot(domain=continent)

chart.coastlines()
chart.land()
chart.borders()
chart.gridlines(draw_labels=False)

chart.subplot_titles("{domain}")

chart.show()
```

### Countries

```python
COUNTRIES = [
    "France",
    "Japan",
    "Madagascar",
    "Cuba",
    "Chile",
    "Australia",
]

chart = Chart(cols=3, rows=math.ceil(len(COUNTRIES)/4))

for country in COUNTRIES:
    chart.add_subplot(domain=country)

chart.coastlines()
chart.land()
chart.borders()
chart.gridlines(draw_labels=False)

chart.subplot_titles("{domain}")

chart.show()
```

### Miscellaneous

```python
DOMAINS = [
    "Alaska",
    "Antarctica",
    "Svalbard",
    "CONUS",
    "Mediterranean",
    "North Atlantic",
]

chart = Chart(cols=3, rows=math.ceil(len(DOMAINS)/4))

for domain in DOMAINS:
    chart.add_subplot(domain=domain)

chart.coastlines()
chart.land()
chart.borders()
chart.gridlines(draw_labels=False)

chart.subplot_titles("{domain}")

chart.show()
```

## Coordinate reference systems

```python
import cartopy.crs as ccrs
```

```python
CRS = [
    ccrs.PlateCarree(),
    ccrs.Robinson(),
    ccrs.Mollweide(),
    ccrs.InterruptedGoodeHomolosine(),
    ccrs.EckertVI(),
    ccrs.NorthPolarStereo(),
]

chart = Chart(cols=3, rows=math.ceil(len(CRS)/4))

for crs in CRS:
    chart.add_subplot(crs=crs)

chart.coastlines()
chart.land()
chart.borders()
chart.gridlines(draw_labels=False)

chart.subplot_titles("{crs}")

chart.show()
```

```python
chart = Chart(cols=3, rows=math.ceil(len(CRS)/4))

for crs in CRS:
    chart.add_subplot(domain="Greenland", crs=crs)

chart.coastlines()
chart.land()
chart.borders()
chart.gridlines(draw_labels=False)

chart.subplot_titles("{domain}\n({crs})")

chart.show()
```
