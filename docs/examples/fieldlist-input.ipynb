```python
import earthkit
```

```python
ens_data = earthkit.data.from_source("file", "/Users/mavj/Downloads/wgust_ens.grib")
data = ens_data.sel(stepRange="72-78")
data.ls()
```

```python
style = earthkit.maps.styles.Contour(
    colors=["#73A3FF", "#0080FF", "#6CA632", "#FF8000", "#FF0A00", "#7B12B3"],
    line_colors="navy",
    levels=range(10, 45, 5),
    extend="max",
)
```

```python
chart = earthkit.maps.Superplot()
chart.plot(data, style=style)
chart.land(color="#B3B3B3")
chart.coastlines()
chart.gridlines(draw_labels=False, xlocs=range(-180, 180, 10), ylocs=range(-180, 180, 10))
chart.borders()
chart.subplot_titles("PF={number}")
chart.subplots[-1].title("CF")

chart.legend(location="bottom")
chart.title()

chart.show()
```
